```python
import sqlite3
import pandas as pd
import numpy as np
from IPython.display import display, HTML
#path = "C:/Users/jxx/AppData/Roaming/fpdb/database/fpdb.db3"
path = "/Users/xxx/.fpdb/database/fpdb.db3"
con = sqlite3.connect(path)

cur = con.cursor()

cur.execute("""SELECT name FROM sqlite_master
            WHERE type='table'
            ORDER BY name;""")

result = cur.fetchall()

print(result)
```

```python
cur.execute("""SELECT * FROM Tourneys limit 5""")
names = list(map(lambda x: x[0], cur.description))
result = cur.fetchall()


df = pd.DataFrame(result, columns=names)

display(HTML(df.to_html()))
```
